```python
selected3= ["close","volume","open","high","low"]
apple1 = pd.read_csv("h:\\Downloads\\apple.csv", sep=",",usecols=selected3)

apple1['volume'] = pd.to_numeric(apple1['volume'].str.replace(',', ''), errors='coerce')

rolling_mean_3 = apple1.rolling(window=3).mean()
rolling_mean_5 = apple1.rolling(window=5).mean()
print("Rolling Mean Window = 3", rolling_mean_3.head(10))
print("Rolling Mean Window = 5", rolling_mean_5.head(10))

```
